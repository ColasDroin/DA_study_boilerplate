### Import modules

```python
# Standard imports
import pandas as pd
import numpy as np
import xtrack as xt
import matplotlib.pyplot as plt
import copy
```

### Load collider

```python
# Define study
collider = xt.Multiline.from_json("/afs/cern.ch/work/c/cdroin/private/example_DA_study_dev/master_study/scans/injection_oct_scan_for_experiment/base_collider/xtrack_0000/collider.json")
```

```python
collider.build_trackers()
```

### Do FMA

```python
fma = collider.lhcb1.get_fma(nemitt_x=2.5e-6, nemitt_y=2.5e-6)
```

```python
fma.plot_fma()
plt.grid(True)
plt.show()
```

### Reverse octupoles and do FMA

```python
collider.vars['i_oct_b1'] = 50.0
collider.vars['i_oct_b2'] = 50.0
```

```python
fma = collider.lhcb1.get_fma(nemitt_x=2.5e-6, nemitt_y=2.5e-6)
```

```python
fma.plot_fma()
plt.grid(True)
plt.show()
```

### Deactivate phase knob and redo

```python
collider.vars['i_oct_b1'] = -50.0
collider.vars['i_oct_b2'] = -50.0

collider.vars["phase_change.b1"] = 0.0
collider.vars["phase_change.b2"] = 0.0

fma = collider.lhcb1.get_fma(nemitt_x=2.5e-6, nemitt_y=2.5e-6)
fma.plot_fma()
plt.grid(True)
plt.show()
```

```python
collider.vars['i_oct_b1'] = +50.0
collider.vars['i_oct_b2'] = +50.0

collider.vars["phase_change.b1"] = 0.0
collider.vars["phase_change.b2"] = 0.0

fma = collider.lhcb1.get_fma(nemitt_x=2.5e-6, nemitt_y=2.5e-6)
fma.plot_fma()
plt.grid(True)
plt.show()
```
